```python
import pandas as pd
df = pd.read_csv("/kaggle/input/datasets/rohitsahoo/sales-forecasting/train.csv")
print(df.head())
```

```python
print("Строк:", df.shape[0])
print("Столбцов:", df.shape[1])
```

```python
df.info()
```

```python
df.describe()
```

```python
print("Общая выручка:", df['Sales'].sum())
```

Общая выручка составила: **2.3 млн** долларов

```python
df['Category'].value_counts()
```

```python
import matplotlib.pyplot as plt

category_sales = df.groupby('Category')['Sales'].sum()
category_sales.plot(kind='bar', color='skyblue')
plt.title('Продажи по категориям')
plt.xlabel('Категория')
plt.ylabel('Выручка')
plt.show()
```

Больше всего продаж в категории: **Technology**  

Меньше всего продаж в категории: **Office Supplies**

```python
# Минимальная цена
print("Минимальная цена:", df['Sales'].min())

# Максимальная цена
print("Максимальная цена:", df['Sales'].max())

# Количество нулевых цен
print("Количество нулевых цен:", (df['Sales'] == 0).sum())

# Количество отрицательных цен
print("Количество отрицательных цен:", (df['Sales'] < 0).sum())

# Количество пустых значений
print("Пустых значений:", df['Sales'].isnull().sum())
```

Аномальных цен не замечено
